# Problema 4

## a)

```python
#Definimos una lista vacía donde guardaremos las 1000 caminatas 
caminatas=[]
""""Utilizamos un for para que vaya llenando la lista, en la cual, cada elemento corresponderá a una caminata,
    es decir, cada elemento será una lista con 30 variables aleatorias """
for i in range(0,1000):
    caminatas.append(random.choices([-1,0,1],weights=(1/3,1/3,1/3),k=30))
```

```python
#Checamos la longitud de caminatas, para ver que efectivamente tenemos 1000
len(caminatas)
```

```python
#Definimos una funcion que nos permita ir encontrando la posicion en cada paso de cada una de las caminatas
def posicion(caminatas,num_pasos):
    posicion=np.zeros((len(caminatas),num_pasos))
    for i in range(0,len(caminatas)):
        for j in range(0,num_pasos):
            posicion[i][j]=sum(caminatas[i][0:j])
    return posicion 
```

```python
#Tomamos solo 20 de las 1000 caminatas que tenemos 
caminatas_20=caminatas[0:20]

#Vamos a definir las posiciones utilizando la función previa, y las 20 caminatas como argumento
posiciones=posicion(caminatas_20,30) #Se toma 30 como número de pasos porque todas las caminatas tienen 30 elementos 
```

```python
#Definimos la desviación estándar utilizando la fórmula vista 
std_p= 2*(np.arange(0,30)/3)**(1/2)
std_n= -2*(np.arange(0,30)/3)**(1/2)
```

```python
#Ahora si vamos a graficar la posición de las 20 caminatas vs el número de pasos 
num_pasos=range(0,30)

for i in range(0,20):
    plt.plot(num_pasos,posiciones[i])
        
#Graficamos también la desviación estándar
plt.plot(num_pasos,std_p, label="Desviación estándar", color="black") 
plt.plot(num_pasos,std_n, color="black") 

plt.xlabel('Número de pasos')
plt.ylabel("Posición")
plt.title('Caminatas aleatorias')
plt.legend(loc="best")
plt.show()
```

## c)

```python
#definimos una función que calcule el promedio de las posiciones en cada paso de cada una de las caminatas
def promedio_posicion(caminatas,num_pasos):
    promedio=np.zeros(num_pasos)
    for i in range(num_pasos):
        suma=0
        for j in range(len(caminatas)):
            suma+=caminatas[j][i]
        promedio[i]=suma/len(caminatas)
    return promedio
```

### 5 caminatas:

```python
#Obtenemos 5 caminatas aleatorias
caminatas_5=[]
for i in range(5):
    caminatas_5.append(random.choices([-1,0,1],weights=(1/3,1/3,1/3),k=30))

#Y les aplicamos la función que calcula el promedio de las posiciones que acabamos de definir    
promedio_5=promedio_posicion(caminatas_5,30)
```

### 100 caminatas:

```python
caminatas_100=[]
for i in range(100):
    caminatas_100.append(random.choices([-1,0,1],weights=(1/3,1/3,1/3),k=30))
    
promedio_100=promedio_posicion(caminatas_100,30)
```

### 500 caminatas

```python
caminatas_500=[]
for i in range(500):
    caminatas_500.append(random.choices([-1,0,1],weights=(1/3,1/3,1/3),k=30))
    
promedio_500=promedio_posicion(caminatas_500,30)
```

### 1000 caminatas

```python
caminatas_1000=[]
for i in range(1000):
    caminatas_1000.append(random.choices([-1,0,1],weights=(1/3,1/3,1/3),k=30))
    
promedio_1000=promedio_posicion(caminatas_1000,30)
```

Finalmente graficamos el arreglo de posiciones promedio en cada caso:

```python
num_pasos=range(30)
plt.scatter(num_pasos,promedio_5,label="5 caminatas")
plt.scatter(num_pasos,promedio_100,label="100 caminatas")
plt.scatter(num_pasos,promedio_500,label="500 caminatas")
plt.scatter(num_pasos,promedio_1000,label="1000 caminatas")
plt.plot(num_pasos,promedio_5)
plt.plot(num_pasos,promedio_100)
plt.plot(num_pasos,promedio_500)
plt.plot(num_pasos,promedio_1000)
plt.plot(num_pasos,np.zeros(30),color="black")

plt.xlabel('Número de pasos')
plt.ylabel("Posición Promedio")
plt.title('Varias caminatas')
plt.legend(loc="best")
plt.show()
```

## e)

```python
#definimos una función que calcule la varianza 
def varianza(posiciones,num_pasos):
    varianza=np.zeros(num_pasos)
    promedio=promedio_posicion(posiciones,num_pasos)
    for i in range(num_pasos):
        suma=0
        for j in range(len(posiciones)):
            suma = suma + pow(posiciones[j][i]-promedio[i],2)
        varianza[i]=suma/len(posiciones)
    return varianza
```

Vamos a utilizar la función de `posición` definida en el inciso a) para obtener las posiciones de cada caminata, y posteriormente la función `varianza` que acabamos de definir para obtener la varianza en cada caso.

```python
#5 caminatas
posiciones_5=posicion(caminatas_5,30)
varianza_5=varianza(posiciones_5,30)
```

```python
#100 caminatas
posiciones_100=posicion(caminatas_100,30)
varianza_100=varianza(posiciones_100,30)
```

```python
#500 caminatas
posiciones_500=posicion(caminatas_500,30)
varianza_500=varianza(posiciones_500,30)
```

```python
#1000 caminatas
posiciones_1000=posicion(caminatas_1000,30)
varianza_1000=varianza(posiciones_1000,30)
```

Finalmente graficamos la varianza vs el número de pasos para cada caso:

```python
num_pasos=np.arange(30)
plt.scatter(num_pasos,varianza_5,label="5 caminatas")
plt.scatter(num_pasos,varianza_100,label="100 caminatas")
plt.scatter(num_pasos,varianza_500,label="500 caminatas")
plt.scatter(num_pasos,varianza_1000,label="1000 caminatas")

plt.plot(num_pasos,varianza_5)
plt.plot(num_pasos,varianza_100)
plt.plot(num_pasos,varianza_500)
plt.plot(num_pasos,varianza_1000)

plt.plot(num_pasos,num_pasos*(2/3),color="black")
plt.xlabel('Número de pasos')
plt.ylabel("Varianza")
plt.title('Varianza vs Número de pasos')
plt.legend(loc="best")
plt.show()
```
