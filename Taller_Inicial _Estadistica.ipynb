```python
#Histogramas
plt.figure(figsize=(10,5))
Arr_datos = np.random.uniform(0,1,1000) #entre mas datos tendera mas a ser una cte
#cada intervalo de datos tiene la misma probabilidad de salir (pq es uniforme)
Arr_bins = np.arange(0,11,0.5)/10 #20 puntos (los valores están divididos entre 10)
#Los bins son las cajitas en donde van los datos del histograma
plt. hist(Arr_datos,Arr_bins)
#Divido una cajita de 0 a 0.5 
#cuantos datos estan en esa cajita
#se repite para todos los intervalos
```

```python
#Normalizar: que fraccion del total de datos salio en cierto intervalo (menor a 1)
plt.figure(figsize=(10,5))
Arr_datos = np.random.uniform(0,1,100000)
Arr_bins = np.arange(0,11,0.5)/10

#Arreglo con ajuste (curva de ajuste)
Arr_fit = np.empty(len(Arr_datos))
for i in range(len(Arr_datos)):
    Arr_fit[i]=1 #Todas las entradas serán 1
#print(Arr_fit)

plt.hist(Arr_datos,Arr_bins, density=True) #norma=1=True #dar densidad de probabilidad
plt.plot(Arr_datos,Arr_fit)
#Se ajustó a una cte 1
#x-datos
#y-fracción
```
