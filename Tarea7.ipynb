# Ejercicio 1

## Supongan que tienen un gas ideal de N moléculas confinado, de modo que la distribución de velocidades está dada por la distribución de Maxwell-Boltzmann. Si $T = 273^o K$ y $m=4.65\times 10^{−26} kg$ (molécula de nitrógeno),

### a) Grafica
$$ n(u) = \left ( \frac{N \lambda_D^3 m^3}{2\pi^2 \hbar^3} \right ) u^2e^{-u^2 m / 2K_b T} $$

### (Recuerden que $\lambda_D = \left ( \frac{2 \pi \hbar^2}{m K_b T} \right) ^{1/2}$ , es decir, $\lambda_D \equiv \lambda_D(T)$).

```python
#Definimos las constantes
T = 273
m = 465E-26
ħ=1.054571E-34
π=3.1416
K_b = 1.380649E-23

λ = ((2*π*(ħ**2))/(m*K_b*T))**0.5
```

```python
#definimos la función de la distribución de velocidades 
def distribucion_vel(u,N):
    u=np.array(u)
    c = (N*(λ**3)*(m**3))/(2*(π**2)*(ħ**3))
    n=c*np.power(u,2)*np.exp((-np.power(u,2)*m)/(2*K_b*T))
    return n
```

```python
u = np.linspace(0,100,100)
N=[1000,5000,10000,20000,50000]
for i in N:
    plt.plot(u,distribucion_vel(u,i),label=f'N = {i}') 
plt.title("Distribución de velocidades")
plt.xlabel("u")
plt.ylabel("n(u)")
plt.legend(loc="best") 
plt.show()
```
