```python
import matplotlib.pyplot as plt
from scipy import optimize,  interpolate, integrate, misc
import seaborn as sns
import numpy as np
import pandas as pd
from sympy import *
```

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9]) #[left, bottom, width, height]
axes_1.set_xlabel('s') #This is how you add labels to the axes
axes_1.set_ylabel('S(s)')
axes_1.set_title('Ecuación trascendental') #Title
βϵnM = .7
βϵnm = 2
s = np.linspace(-5,5,100)
SM = np.tanh(βϵnM*s)
Sm = np.tanh(βϵnm*s)
axes_1.plot(s,SM,label='β<1') #To plot the axes, and a label generated
axes_1.plot(s,Sm,label='β>1') 
axes_1.plot(s,s,label='S') 
axes_1.legend(loc=0) #Legend

axes_1.set_xlim(-1.5, 1.5)
axes_1.set_ylim(-1.5, 1.5)
#plt.show()
```

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9]) #[left, bottom, width, height]
axes_1.set_xlabel('x') #This is how you add labels to the axes
axes_1.set_ylabel('F/U0')
axes_1.set_title('Ecuación trascendental') #Title
kbT= 0
kb= 8.617333262*(10**(-5))
x = np.linspace(-10,10,2000)
a = (1-x)/2
b = (1+x)/2
F =-(x**2)+ (kbT)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F,label='kbT=0') #To plot the axes, and a label generated
kbT1=1
F1 =-(x**2)+ (kbT1)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F1,label='kbT=1')
kbT2=2
F2 =-(x**2)+ (kbT2)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F2,label='kbT=2')
kbT3=3
F3 =-(x**2)+ (kbT3)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F3,label='kbT=3')
axes_1.legend(loc=0) #Legend

axes_1.set_xlim(-1, 1)
#axes_1.set_ylim(-1, 1)
#plt.show()
```
