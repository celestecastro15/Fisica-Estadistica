# Ejercicio 5

## Usa las soluciones $X(t) = N_{0,x}^t (0,1) \sqrt{\delta^2 t}$ y $Y(t) = N_{0,y}^t (0,1) \sqrt{\delta^2 t}$  y el método en la sección 6.4 (del libro) para generar y graficar el camino de una partícula Browniana en el plano x-y. Supón que las normales unitarias $N_{0,x}^t (0,1)$ y $N_{0,y}^t (0,1)$ son estadísticamente independientes.

Vamos a encontrar la solución a la posición de la partícula avanzando numéricamente mediante la siguiente ecuación de actualización:

$$ X (t + \Delta t ) = x (t) + N_t^{t+\Delta t} (0,1) \sqrt{\delta^2 \Delta t} 
$$

en donde, $\delta$ representa el tamaño del paso y está dada por: $\delta^2 = \left (\frac{\Delta x^2}{\Delta t} \right )$

Este método es el llamado Monte Carlo, el cual es una secuencia de tales actualizaciones con la realización de la posición actualizada $x(t + \delta t)$ en el final de cada paso de tiempo, utilizado como la posición inicial $x(t)$ al comienzo de la
siguiente. A continuación vamos a programar lo anterior:

```python
i = 1000 #número de iteraciones
tiempo=np.array(range(i)) #intervalo de tiempo, en donde el paso va a ser de 1, i.e. Δt=1
Δt=1

#Como tanto el arreglo que definirá a X y Y, como el del tiempo van a estar dados en pasos de 1 , entonces δ^2=1
δ=1

#Empezamos definiendo para X(t)
X = np.zeros(i)
for t in range(i-1):
    X[t+Δt] = X[t] + np.random.normal(0,1) * np.sqrt(δ*Δt)
plt.plot(tiempo,X, color="#33FDFF", label="X(t)")

#Repetimos para Y(t)
Y = np.zeros(i)
for t in range(i-1):
    Y[t+Δt] = Y[t] + np.random.normal(0,1) * np.sqrt(δ*Δt)
plt.plot(tiempo,Y, color="#33FF5D",label="Y(t)")

#Graficamos X(t) y Y(t) vs tiempo
plt.title("Solución para X(t) y Y(t) por el método Monte Carlo")
plt.xlabel('t')
plt.ylabel('Solución')
plt.legend(loc = "best")
plt.show()
```

Ahora sí, como ya tenemos programado todo lo necesario, procedemos a graficar $X(t)$ vs $Y(t)$ para obtener la trayectoria de la partícula en el plano x-y:

```python
plt.plot(X,Y,color="red",label="trayectoria")
plt.scatter(X[0],Y[0],color="blue",label="inicio")
plt.scatter(X[-1],Y[-1],color="green",label="final")
plt.title("Partícula browniana en el plano XY")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc="best")
plt.show()
```
