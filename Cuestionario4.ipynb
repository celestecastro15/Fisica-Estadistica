# Problema 1

## a) Esquemáticamente, muestra cómo cambia la probabilidad de que una partícula en dicho sistema tenga velocidad $v$ como función del tiempo.

$$ p(v,t) = \frac{\exp \left( \frac{- ( v - \nu_0 e^{-\gamma t})^2}{2 \frac{\beta^2}{2 \gamma} ( 1 -e^{-2\gamma t})}  \right )}{\sqrt{ 2 \pi \left ( \frac{\beta^2}{2 \gamma} \right ) (1-e^{-2\gamma t})}} $$

Definimos primero una función que nos va a calcular la distribución de probabilidad para una velocidad $v$, como función de los coeficientes $\gamma$ y $\beta$, la velocidad incial $v_0$ y un intervalo de tiempo:

```julia
function probabilidad(v,γ,β,v0,tiempo)
    γ=1
    β=1
    p=[]
    for t in tiempo
        aux1 = exp( (-(v-v0*exp(-γ*t))^2) /( (β^2/γ )*(1-exp(-2*γ*t)) ))
        aux2 = ( 2*π*((β^2)/(2*γ))*(1-exp(-2*γ*t)) )^(-1/2)
        valor = aux1 * aux2
        push!(p,valor)
    end
    return p
end
```

Y graficamos para un intervalo de tiempo dado y para varias velocidades:

```julia
γ=1
β=1
v0=0
tiempo = collect(0:0.01:10)
velocidades=[0.5,1,1.5,2,3,4,5,10]
plot(title="Probabilidad de que la partícula tenga velocidad v",xlabel="t",ylabel="p(v,t)")
for v in velocidades 
     plot!(tiempo, probabilidad(v,γ,β,v0,tiempo),label="v = $(v)")
end
plot!()
```
