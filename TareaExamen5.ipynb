### e)

Haciendo el término $k_b$ del denominador igual a 1:

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('x') 
axes_1.set_ylabel('F/Uₒ')
axes_1.set_title('Energía libre')

x = np.linspace(-0.9,0.9,2000)
a = (1-x)/2
b = (1+x)/2

kbT= 0
F =-(x**2)+ (kbT)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F,label='kbT=0',color="#33FF73") 

kbT1=1
F1 =-(x**2)+ (kbT1)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F1,label='kbT=1',color="#33D2FF")

kbT2=2
F2 =-(x**2)+ (kbT2)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F2,label='kbT=2',color="#9D33FF")

kbT3=3
F3 =-(x**2)+ (kbT3)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F3,label='kbT=3',color="#FF33A8")

axes_1.legend(loc="best")
axes_1.set_xlim(-1, 1)
```

Dejando el término $k_b$ en el denominador:

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9]) 
axes_1.set_xlabel('x') 
axes_1.set_ylabel('F/Uₒ')
axes_1.set_title('Energía libre')

kb= 8.617333262*(10**(-5))
x = np.linspace(-0.9,0.9,2000)
a = (1-x)/2
b = (1+x)/2

kbT= 0
F =-(x**2)+ (kbT/kb)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F,label='kbT=0') 

kbT1=1
F1 =-(x**2)+ (kbT1/kb)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F1,label='kbT=1')

kbT2=2
F2 =-(x**2)+ (kbT2/kb)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F2,label='kbT=2')

kbT3=3
F3 =-(x**2)+ (kbT3/kb)*(a*np.log(a) + b*np.log(b) )
axes_1.plot(x,F3,label='kbT=3')

axes_1.legend(loc="best") 
axes_1.set_xlim(-1, 1)
```
